# 2-3. Row/column 선택·추가·삭제
## 1. row/column 선택
(1) row 선택하기: **DataFrame[n:m]** (n번째 행부터 m 직전 행까지)\
(2) column 선택하기:
> **DataFrame['컬럼명']** (하나의 컬럼을 Series 형식으로 출력)\
> **DataFrame[['컬럼명1', '컬럼명2']]** (여러개의 컬럼을 DataFrame 형식으로 출력)

(3) row와 column 선택하기: **iloc[row, column]**, **loc[row, column]**
> - iloc: 정수로 된 위치 기반 인덱스로 행과 열을 선택할 때
> - loc: 레이블(인덱스의 이름 혹은 컬럼명)로 행과 열을 선택할 때, loc에서 슬라이싱 start:end를 사용할 때는 **end 포함** 주의

```python
from sklearn.datasets import load_iris
import pandas as pd
iris = load_iris()
iris = pd.DataFrame(iris.data, columns=iris.feature_names)
iris[1:4] # 객체의 인텓스는 0부터 시작
```

```python
# 선택한 값 변경하기
score = pd.DataFrame({'국어':[100,80],'수학':[75,90],'영어':[90,95]},index=['장화','홍련'])
score.loc['홍련','영어'] = 100
score['국어'] = score['국어'] - 5
score
```

## 2. row/column 추가
(1) row 추가: **concat([기존,추가], ignore_index=False)** (인수로 Series, DataFrame, Dictionary)\
(2) column 추가: 단일 값 또는 데이터프레임 행의 수와 같은 길이의 배열을 추가할 수 있음\
(3) row/column 삭제: **drop()** (행과 열 동시에 삭제하는 것은 불가능)

```python
# 행 추가
new_students = pd.DataFrame({'국어':[100,80],'수학':[75,90],'영어':[90,95]}, index=['콩쥐','팥쥐'])
score = pd.concat([score,new_students], ignore_index=False)
new_student1 = pd.Series({'국어':75,'수학':80,'영어':80}, name='해님')
score = pd.concat([score,pd.DataFrame(new_student1).transpose()])
score
```

```python
# 열 추가
science = [80, 70, 90, 85, 75]
score['과학'] = science # 배열
score['학년'] = 1 # 단일 값
score
```

```python
score['과학'] = score['과학'] + 5
score['총점'] = score['국어'] + score['수학'] + score['영어'] + score['과학']
score
```

```python
# 행 또는 열 삭제
score.drop('장화', inplace=True)
score.drop(columns = ['과학', '학년', '총점'], inplace=True)
score
```
